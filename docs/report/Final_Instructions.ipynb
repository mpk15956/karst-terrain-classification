## III. Project Examples

If you are hunting for ideas, consider these categories as starters:

### Type 1: Model Comparison for the Same Task
Compare the performance of **different models** on the **same task**.

**Example:** *Urban Tree Species Classification from Street View Imagery*
- Task: Classify tree species in urban environments using Google Street View images
- Models: Compare ResNet-50, EfficientNet, and Vision Transformer (ViT)

### Type 2: Performance Improvement with the Same Model
Use the **same model** for a task and explore **multiple approaches** to improve its performance.

**Example:** *Improving Geocoding Accuracy for Historical Place Names*
- Model: Use a single LLM (e.g., GPT-4 or LLaMA)
- Task: Geocode historical place names from 19th-century documents
- Improvements to test:
  - Baseline: Zero-shot prompting
  - Method 1: Few-shot prompting with historical examples

### Type 3: Adapting a Model to a Non-Original Task
Apply a model to a task it was **not originally designed for** and explore ways to improve its performance.

**Example:** *Predicting Flood Risk from Social Media Images*
- Model: DETR (originally designed for object detection)
- New task: Assess flood severity and water level from user-posted images
- Adaptations:
  - Fine-tune on labeled flood imagery dataset
  - Modify model to predict water level regression instead of just detection
  - Add temporal context from multiple images

### Type 4: AI vs. Non-AI Solution Comparison
Compare an **AI model's performance** with a **traditional non-AI approach**. This is probably the most challenging type as you need to implement both AI solution and the conventional solution.

**Example:** *Land Cover Classification: Deep Learning vs. Traditional Remote Sensing*
- Task: Classify land cover types from satellite imagery
- AI approach: Fine-tuned ResNet on labeled satellite patches
- Non-AI approach: Traditional NDVI thresholding and unsupervised clustering (k-means)

### Example Projects to **Avoid**

- No AI Model Used: A project that does not incorporate any pre-trained AI model.
- Simply applying one model version with no analysis: No exploration or tweaking of its behavior, no systematic evaluation of performance, no comparison or experimentation, or just running a model once and presenting the output. These are considered too light-weight and should be more carefully developed.
